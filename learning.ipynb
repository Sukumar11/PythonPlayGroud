```python
#classes and objects
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."
    
person1 = Person("Sukku", 25)
print(person1.greet())

class phone:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def get_info(self):
        return f"Phone: {self.brand} {self.model}"
    
    def make_call(self, number):
        return f"Calling {number} from {self.brand} {self.model}..."
    
phone1 = phone("Apple", "iPhone 13")
print(phone1.get_info())
print(phone1.make_call("123-456-7890"))
```

```python
#oops concepts

class Animal:

    def __init__(self, name, species):
        self.name = name
        self.species = species

    def make_sound(self):
        if self.species == "dog":
            return "Woof!"
        elif self.species == "cat":
            return "Meow!"
        else:
            return "Unknown sound"  

class Dog(Animal):
    def __init__(self, name):
        super().__init__(name, "dog")   

class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, "cat")   

dog1 = Dog("Buddy")
cat1 = Cat("Whiskers")
print(dog1.name)  # Output: Buddy
print(cat1.name)  # Output: Whiskers    
print(dog1.make_sound())  # Output: Woof!
print(cat1.make_sound())  # Output: Meow!
```
